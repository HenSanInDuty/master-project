```python
# Need cluster document
documents = []
for file in file_origin:
    document = {
        'topic':'',
        'content':''
    }
    with open(file, "r", encoding="utf-8") as f:
        topic = ''
        all_sen = f.read().split("\n")
        key = all_sen[2].split(" ")[1].split("/")[2]
        link = all_sen[2].split(" ")[1]
        index = topic_index[f'{key}']
        if index != '':
            if isinstance(index, int):
                topic = link.split("/")[index]
            else:
                topic = index
        else:
            continue
        for i in range(8, len(all_sen)):
            document['content'] += all_sen[i]
        document['topic'] = topic.lower()
        documents.append(document)
```

```python
# Need test cluster
test_documents = []
for cluster in cluster_dir_origin:
    test_document = {}
    # Đọc origin
    file_origin_path = data_vims_origin_dir + f"\\{cluster}\\original"
    file_origin = []
    document_origin = []
    
    for f in os.listdir(file_origin_path):
        if f == '.DS_Store':
            continue
        join_file = os.path.join(file_origin_path, f)
        if os.path.isfile(join_file):
            file_origin.append(join_file)
            
    for file in file_origin:
        document = {
            'topic':'',
            'content':''
        }
        
        with open(file, "r", encoding="utf-8") as f:
            topic = ''
            all_sen = f.read().split("\n")
            key = all_sen[2].split(" ")[1].split("/")[2]
            link = all_sen[2].split(" ")[1]
            index = topic_index[f'{key}']
            if isinstance(index, int):
                topic = link.split("/")[index]
            else:
                topic = index
            for i in range(8, len(all_sen)):
                document['content'] += all_sen[i]
            document['topic'] = topic.lower()
            document_origin.append(document)
            
    # Đọc summary
    file_summary_path = data_vims_sum_dir + f"\\{cluster}"
    summary_content = []
    file_summary = []
    # Liệt kê các file trong thư mục summary
    for f in os.listdir(file_summary_path):
        if f == '.DS_Store':
            continue
        join_file = os.path.join(file_summary_path, f)
        if os.path.isfile(join_file):
            file_summary.append(join_file)
            
    for file in file_summary:
        with open(file, "r", encoding="utf-8") as f:
            summary = ''.join(f.read().split("\n"))
            summary_content.append(summary)
    test_documents.append({'documents': document_origin, 'summary_content': summary_content}) 
```
